### To sort the list in ascending or descending

```python
def BubbleSort(arr):         #Ascending
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr=list(map(int,input("Enter the list: ").split(",")))
print("List after sorting:",BubbleSort(arr))
```

```python
def BubbleSort(arr):       #Descending
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]<arr[j+1]:
                arr[j+1],arr[j]=arr[j],arr[j+1]
    return arr

arr=list(map(int,input("Enter the list: ").split(",")))
print("List after sorting:",BubbleSort(arr))
```

# Manipulate the lists

```python
class list_op:
    def list_op_func(self):
        N = int(input("Enter the number of operations you want to perform: "))
        lists = []

        for i in range(N):
            try:
                command = input(f"Enter the command {i+1}: ").split()
                command[0] = command[0].casefold()

                if command[0] == "insert":
                    lists.insert(int(command[1]), int(command[2]))
                    print("List after insertion:", lists)

                elif command[0] == "append":
                    lists.append(int(command[1]))
                    print("List after appending:", lists)

                elif command[0] == "remove":
                    lists.remove(int(command[1]))
                    print("List after removing:", lists)

                elif command[0] == "print":
                    print("List:", lists)

                elif command[0] == "pop":
                    lists.pop()
                    print("List after popping:", lists)

                elif command[0] == "reverse":
                    lists.reverse()
                    print("List after reversing:", lists)

                elif command[0] == "sort":
                    try:
                        Nsort = input("Enter 'Asc' or 'Desc': ").casefold()

                        if Nsort == "asc":
                            lists.sort()
                            print("List after sorting Asc:", lists)

                        elif Nsort == "desc":
                            lists.sort(reverse=True)
                            print("List after sorting Desc:", lists)

                        else:
                            print("!! Enter only Asc or Desc !!")

                    except Exception:
                        print("!! Sorting error !!")

                else:
                    print("!! Invalid command !!")

            except Exception as e:
                print("!! Enter valid command and arguments !!")


if __name__ == "__main__":
    obj = list_op()
    obj.list_op_func()
```

### To swap the case

```python
def swap_str(s):
    s = s.swapcase()
    return print_str(s)

def print_str(s):
    print("Swapped string:",s)

if __name__ == "__main__":
    s = input("Enter the string:")
    swap_str(s)
```

### To split and join the string

```python
def split_join(string):
    string = string.split(" ")
    string = "-".join(string)
    return string

if __name__ == "__main__":
    string = input("Enter the string:")
    result = split_join(string)
    print("String after joining:", result)
```

### Try except else finally

```python
try:
    # Code that might cause an error
    numerator = 10
    denominator = 2
    result = numerator / denominator
except ZeroDivisionError:
    # Code that runs if a ZeroDivisionError occurs
    print("Error: Cannot divide by zero!")
else:
    # Code that runs only if the 'try' block was successful
    print(f"The result is {result}")
finally:
    # Code that always runs, regardless of an exception
    print("Execution finished.")

```

### Length of similar string

```python
def count_substring(string, sub_string):
    count = 0
    sub_len = len(sub_string)
    for i in range(len(string)- sub_len +1):
        if string[i:i+sub_len] == sub_string:
            count+=1
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```
