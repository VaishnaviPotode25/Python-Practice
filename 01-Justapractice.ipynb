# Factorial

```python
def factorial(n):

    fact = 1
    if n==0 or n==1:
        return 1
    else:
        fact = n * factorial(n-1)
        return fact
        
n = int(input("Enter the number:"))
factorial(n)
```

# Palindrome

```python
def palindromeCheck(num):
    copy = num
    rev = 0
    while num>0:
        rev = rev * 10 + num % 10
        num = num // 10
    if copy == rev:
        print(f"{copy} is palindrome")
    else:
        print(f"{copy} is not palindrome")

num=int(input("Enter the number:"))
palindromeCheck(num)
```

```python
def palindromeCheck(str):
    str=str.casefold()
    check=""
    for i in range(len(str)-1,-1,-1):
        check = check + str[i]
    if check == str:
        print(f"{str} is palindrome")
    else:
        print(f"{str} is not palindrome")

str=input("Enter the string:")
palindromeCheck(str)
```

# Fibonacci series

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
n=int(input("Enter the number:"))
for i in range(n):
   print(fibonacci(i),end=" ")
```

# Zig Zag Conversion

```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1 or numRows >= len(s):
            return s
        rows = [""] * numRows
        currentRow = 0
        direction = -1

        for char in s:
            rows[currentRow] += char
            if currentRow == 0 or currentRow == numRows - 1:
                direction *= -1
            currentRow += direction
            
        return "".join(rows)

sol=Solution()
sol.convert("PAYPALISHIRING",3)
        
```

# Sum of index

```python
class SumIndex:
    def target_sum(self, nums , target):
        seen={}
        for i, num in enumerate(nums):
            need=target-num
            if need in seen:
                return [seen[need],i]
            seen[num]=i
        return []

nums=list(map(int,input("Enter the list(comma separated):").split(",")))
target=int(input("Enter the target:"))
si=SumIndex()
si.target_sum(nums,target)
```

# N repeated elements

```python
class Solution:
    def repeatnum(self,nums):
        
        n=len(nums)-1
        if nums[0]==nums[n]:   
            return nums[0]
        for i in range(n):
            if nums[i]==nums[i+ 1] or nums[i]==nums[i+ 2]:
                return nums[i]

        return nums[0]

nums=list(map(int,input("Enter the list(comma separated):").split(",")))
sol=Solution()
sol.repeatnum(nums)
```

```python
class Solution:
    def repeatnum(self,nums):
        seen=set()
        for i in nums:
           if i not in seen:
              seen.add(i)
           else:
              return i

nums=list(map(int,input("Enter the list(comma separated):").split(",")))
sol=Solution()
sol.repeatnum(nums)
```

# Inheritance

```python
class Car:
    def __init__(self, brand, speed=0):
        self.__brand = brand  # Private attribute
        self.__speed = speed  # Private attribute

    def accelerate(self, increase):
        self.__speed += increase

    def get_speed(self):
        return self.__speed

    def get_brand(self):
        return self.__brand
```

```python
car=Car("Tata")
car.accelerate(40)
car.get_speed()
```

### To find the smallest no.in list

```python
def SmallestNum(array_list):
    smallest_num=array_list[0]
    for num in array_list:
        if num<smallest_num:
            smallest_num=num
    return smallest_num

array_list=list(map(int,input("Enter the list: ").split(",")))
SmallestNum(array_list)
```
