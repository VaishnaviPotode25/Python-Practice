# Data Cleaning

```python
df['sales'] = pd.to_numeric(df['sales'], errors="coerce")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.info()
```

# Feature Selection

```python
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days
df['profit_margin'] = (df['profit'] / df['sales'])
```

```python
df[['sales','profit','discount','shipping_days','profit_margin']].head()
```

# Univariate Analysis

## 1.Sales Distribution

```python
plt.figure()
plt.hist(df['sales'].dropna() ,bins=50,color='red')
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Sales Distribution")
plt.show()
```

## 2.Profit Distribution

```python
plt.figure()
plt.hist(df['profit'].dropna() ,bins=50,color='b')
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit Distribution")
plt.show()
```

## 3.Discount Distribution

```python
plt.figure()
plt.hist(df['discount'].dropna() ,bins=50,color='orange')
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.title("Discount Distribution")
plt.show()
```

# Bivariate Analysis

## 1.Dicount vs Profit

```python
plt.figure()
plt.scatter(df['discount'],df['profit'],alpha=0.3,color="b")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Dicount VS Profit")
plt.show()
```

```python
plt.figure()
plt.scatter(df['sales'],df['profit'],alpha=0.3,color="b")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales VS Profit")
plt.show()
```
